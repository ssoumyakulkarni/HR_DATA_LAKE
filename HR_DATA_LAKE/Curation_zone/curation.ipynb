```python
def email_val(data):
    numOfRows = data.shape[0]
#     print(numOfRows)
    for i in range(0,numOfRows):
    #     print(cap.email[i])
        address = data.email[i]
        bool_result = is_email(address)
        detailed_result = is_email(address, diagnose=True)
        if(bool_result == True):
            data.at[i,'valid_email']= "YES"
        else:
            data.at[i,'valid_email'] = "NO"
            
def emp_visitor(data, eos):
    #check using eos data
    record = data['email'].isin(eos['Email_Id'])
    data['employee'] = record
    
    #check using domain name
    numOfRows = data.shape[0]
    for i in range(0,numOfRows):
        address = data.email[i]
#         print(address)
        if(address.split('@')[1] == "aroha.co.in"):
            data.at[i,'employee'] = True
```

# Grouping of data 


```python
cap_g = cap.groupby(['First_Name', 'Last_Name','email','phone', 'Skill_Level', 'Topic_Name','employee'], sort=False).sum().reset_index()
cap_g.to_csv('capability.csv', index=False)
cap_g.head()
```

```python
ilearn_g = ilearn.groupby(['first_name', 'last_name','email', 'course','employee'], sort=False).agg({'duration':'sum'}).reset_index()
ilearn_g.to_csv('ilearn.csv', index=False)
ilearn_g.head()
```

```python
pet_g = pet.groupby(['name','email','phone_no', 'primary_skills','employee'], sort=False).agg({'noOfException':'sum', 'noOfScenario':'sum','noOfSqlStr':'sum','productivity':'mean'}).reset_index()
pet_g.to_csv('pet.csv', index=False)
pet_g.head()
```

# Creating Master data

```python
unique_mail = (ilearn_g['email'].append(pet_g['email']).append(cap_g['email'])).unique()
```

```python
print(len(unique_mail))
```

```python
merge = pd.DataFrame(columns=['email', 'phone', 'capabilityUserInd', 'ilearnUserInd', 'productivityUserInd'])
def phone(unique_mail, ilearn_g, pet_g, cap_g):
    if unique_mail in cap_g['email'].values:
        return cap_g.loc[cap_g['email'] == unique_mail, 'phone'].iloc[0]
    if unique_mail in pet_g['email'].values:
        return pet_g.loc[pet_g['email'] == unique_mail, 'phone_no'].iloc[0]
#     if unique_mail[i] in ilearn_g['email'].values:
#         return ilearn_g.loc[ilearn_g['email'] == unique_mail[i], 'phone'].iloc[0]
```

```python
for i in range(len(unique_mail)):
    merge.at[i,'email'] = unique_mail[i]
    ph = phone(unique_mail[i], ilearn_g, pet_g, cap_g)
    merge.at[i,'phone'] = ph  
    merge.at[i,'capabilityUserInd'] = "YES" if unique_mail[i] in cap_g['email'].values else "NO"
    merge.at[i,'ilearnUserInd'] = "YES" if unique_mail[i] in ilearn_g['email'].values else "NO"
    merge.at[i,'productivityUserInd'] = "YES" if unique_mail[i] in pet_g['email'].values else "NO"
```

```python
merge.to_csv("KIPuserList.csv")
merge.head()
```
